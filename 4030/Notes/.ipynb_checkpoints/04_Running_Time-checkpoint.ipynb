```python
def clean_data(text):
    # split and strip text into list of lines
    # go through each line, split it into a list of things
    # append that list to output
    lines = text.strip().split('\n')
    output = []
    for line in lines:
        things = line.split(',')
        output.append(things)
        
    return output

def compute_avg_hp(data):
    # go through each item in the data, skipping the first item (header)
    # pick out the horse power from the current line
    # add the horsepower to a current sum
    # return average
    sum = 0
    count = 0
    for i in range(1, len(data)):
        if data[i][3] != '':
            hp = float(data[i][3])
            sum += hp
            count += 1

    return sum / count
    
        
        
        
            
    
    
    
```

+ Key concepts: running time functions, run time complexity.

```python
def prob1(L):
    s = 0
    for x in L:
        s += x
    s = s * 5
    return s
```

The running time function of a program specifires the number of steps the program takes. 

Definition: T(n) is the running time equation of prob1, where n is the number of items on the input list. 

n is what we call input size. 

The running time 

T(21) is the running time of the program when the input has 25 items. 

T(n) is the running time of this program when the input has n items.

How do we detriming T(n)? Answer: we sum up all the steps in the program. 

```python
def prob1(L): #T(n)
    s = 0 # 1 step (a1 steps)
    for x in L: # steps (L has n items)
        s += x # each iteration takes: b steps
    s = s * 5 # 1 step (a2 steps)
    return s # a3 steps
```

T(n) = a + b

a = a1 + a2 + a3

if n = 10, T(10) = a + 10b

This running time function linearly dependent on n.

### O (or Big O)

O is a measurement of complexity.

O is a way to specify upper bounds of (resource) functions.

#### A running time complexity of O(n) means that the program

O(n) means: T(n) ≤ c * n

Upper bound estimation is very useful for complex programs.

T(n) = a + b * n ≤ a * n + b * n for n > 1

Im upperbounding T(n), specifically the term 'a'.

T(n) ≤ (a + b) * n

In other words, T(n) ≤ c * n.

Therefore, T(n) is in O(n).

O(n) is a set of functions.

Revising the complexity of prob1.

We've established that T(n) ≤ c * n ≤ c * n^2 for all n > 1

Therefor by definition, $T(n) \in 0(n)$ means the upper bound of this runing time function is a constant times n.

$T(n) \in 0(n^2)$ means the upper bound of this running time function is a constant times $n^2$

# True or False?

+ $5n^3 + 10 \in O(n^2)$ FALSE

+ $5n^3 + 10 \in O(n^3)$ TRUE

We can show this using the definition of O. (upcoming assignments)

$5n^3 + 10 ≤ 5n^3 + 10n^3 = 15n^3$ , for all n > 1

I have chosen the constant c to be 15. With this choice, we've shown that  $5n^3 + 10 \in O(n^3)$.

$T(n) \in O(n^3)$ means T(n) ≤ c * n^3 for some constant c and when n is greater than some value.

If you want to prove that T(n) $\in O(n^3)$, you have to identify a specific constant c.

# An example

Show that $n^4 + 5 \in O(n^7)$

AnswerL

$n^4 + 5 ≤ n^7 + 5n^7 = 6n^7$, for all n < 1

With the choice of c = 6,$n^4 + 5 ≤ c * n^7$.

$n^7$ is not a tight upper bound, but its still an upper bound.

## Definition of O:

$T(n) \in O(f(n)) if T(n) ≤ c * f(n) for some specific value of c and for n larger than a number.

This means f(n) is an upper bound of T(n).

```python
#
# L is a list of numbers.
# We dont know in advance how many numbers, so well just say 'n'.
# 
def prob2(L):                          # T(n)
    s = 0                              # a1 steps
    for i in range(len(L)):            # n iterations
        for j in range(i + 3, len(L)): # n iterations
            s += L[i]*L[j]             # b steps
    return s                           # a2 steps
```

```python
$T(n) = a + b * n^2$ is no longer correct because the second loop does not have n iterations.
```
